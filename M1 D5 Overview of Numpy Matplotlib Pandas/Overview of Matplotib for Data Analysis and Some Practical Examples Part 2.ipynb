# Matplotlib

<p>Matplotlib is probably the single most used Python package for 2D-graphics. It provides both a very quick way to visualize data from Python and publication-quality figures in many formats.</p>

*matplotlib.pyplot* is a collection of command style functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc. 

```python
#1

import matplotlib.pyplot as plt

x = [1,2,5,6,7]
y = [4,8,12,16,24]
plt.title('Dummy Graph')
plt.ylabel('BIG NUM')
plt.xlabel('Num')
#plt.legend? #Assignment
plt.plot(x,y)

plt.show()
```

```python
#2

import matplotlib.pyplot as plt

t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2*np.pi*t)

plt.plot(t, s)

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('About as simple as it gets, folks')
plt.grid(True)
plt.show()

```

```python
# Pie chart

import matplotlib.pyplot as plt

#n = 20
#Z = np.random.uniform(0,1,n)
Z = np.array([300,200,250,210,400])
labs=['A','B','C','D','E']
plt.pie(Z,explode=(1,.1,.12,.1,.12),labels=labs)
plt.gca().set_aspect('equal') # Assignment

plt.show() 

# Name Segments of pi chart
```

```python
# Let See 3D plots

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')

plt.show()
```
