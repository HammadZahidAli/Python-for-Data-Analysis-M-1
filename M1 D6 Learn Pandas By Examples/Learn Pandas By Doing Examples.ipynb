```python
# 2.4 What's the most common complaint type?
complaints_df['Complaint Type'].value_counts()
```

```python
ctv = complaints_df['Complaint Type'].value_counts()

topten_ctv = ctv[:10]
topten_ctv
```

```python
%matplotlib inline 



topten_ctv.plot(kind='bar',figsize=(10,5))


```
